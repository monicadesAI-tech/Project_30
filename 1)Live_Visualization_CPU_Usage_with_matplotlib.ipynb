```python
import time 
import psutil
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
plt.rcParams['animation.html'] = 'jshtml'
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
fig.show()
```

```python
psutil.cpu_percent()
```

```python
i = 0
x, y = [], []
while True:
    x.append(i)
    y.append(psutil.cpu_percent())
    
    ax.plot(x,y,color='b')
    
    fig.canvas.draw()
    
    ax.set_xlim(left=max(0,i-50),right = i+50)  # To get live data only and not old, if you want old data also then remove this line
    
    time.sleep(0.1)
    i += 1
    
```

```python
plt.close()
```
